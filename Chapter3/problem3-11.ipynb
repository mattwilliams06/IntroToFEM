```python
from lagrange import Lagrange
```

```python
x = [0.0,1.0,2.0,3.0,4.0]
y = [1.0,2.7,5.8,6.6,7.5]
li = Lagrange()
yi,xi = li(x,y,True)
```

```python
plt.scatter(x,y,c='r')
plt.plot(xi,yi)
```

```python
def center_diff(x,y):
    dx = x[1]-x[0]
    dydx = (y[2:] - y[:-2])/(2*dx)
    return dydx, x[1:-1]
```

```python
dy, xdiff = center_diff(xi,yi)
```

```python
plt.plot(xdiff,dy)
```

```python
x = [0.0,1.0,2.0,3.0,4.0]
y = [1.0,2.7,5.8,6.6,7.5] + np.random.randn(len(y))
li = Lagrange()
yi,xi = li(x,y,True)
plt.scatter(x,y,c='r')
plt.plot(xi,yi)
```

```python
dy, xdiff = center_diff(xi,yi)
plt.plot(xdiff,dy)
```
