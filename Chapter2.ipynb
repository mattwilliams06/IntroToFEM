# Problem 2.9

```python
x = np.linspace(0,5,501)
y = np.linspace(0,5,501)
X, Y = np.meshgrid(x,y)
u = lambda x,y: np.sqrt(x**2+(y-3)**2)
plt.contourf(X,Y,u(X,Y))
```

```python
Z = np.zeros_like(X)
for row in range(X.shape[0]):
    for col in range(X.shape[1]):
        r = np.sqrt(X[row,col]**2+Y[row,col]**2)
        if r <= 2:
            Z[row,col] = -1.0
        else:
            Z[row,col] = u(X[row,col],Y[row,col])
```

```python
levels = np.linspace(0,6,13)
plt.contourf(X,Y,Z,levels=levels)
# levels = 0.0
# plt.contourf(X,Y,Z,level=levels,colors='w')
plt.colorbar()
```
